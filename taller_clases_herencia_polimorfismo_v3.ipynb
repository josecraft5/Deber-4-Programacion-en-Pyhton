---


> 💡 **Concepto clave**
>
> Una **clase** es como una ficha técnica de un equipo. La ficha describe qué datos tiene (atributos) y qué puede hacer (métodos). Cuando creas un equipo real a partir de esa ficha, obtienes un **objeto**. Puedes crear 10 reactores distintos con la misma ficha.


### Las cuatro piezas de toda clase

| Pieza | Qué hace |
|-------|----------|
| **Atributos** | Datos del objeto. Ej: `nombre`, `temperatura` |
| **`__init__`** | Constructor: se ejecuta al crear el objeto, guarda los datos |
| **Métodos** | Funciones del objeto. Definen su comportamiento |
| **`self`** | Referencia al objeto actual — "este reactor en particular" |

> ⚠️ Todo método de una clase **siempre** recibe `self` como primer parámetro.
> Es el error más común al comenzar.

### Prueba de concepto — completa y ejecuta:

>

> - ✅ `class Equipo_Prueba:` — estructura de la clase
> - ✅ `def __init__(self, nombre):` — constructor ya definido
> - ✅ `def describir(self):` — método ya definido
>
> Debes **completar** — reemplaza cada `___`:
> - ✏️ `___` `self.nombre = ___` — asigna el parámetro `nombre` al objeto
> - ✏️ `___` el texto del `print` en `describir()` — usa `self.nombre`
> - ✏️ `___` el argumento al instanciar: el nombre del equipo de prueba

> y puede estar **encendido o apagado**.
> Empieza por ahí. Esa es la base de todo el sistema."*

Miras el cuaderno. Hay una tabla de equipos registrados:

| Código | Nombre | Estado inicial |
|--------|--------|----------------|
| EQ-001 | Mezclador Central | INACTIVO |
| EQ-002 | Centrífuga | INACTIVO |
| RX-001 | Reactor-A | INACTIVO |

Empiezas con la clase base que todos compartirán.

```python
# ── Estado del sistema hasta aquí ──────────────────────

```

> **Cómo leer esta celda:**
>
> Ya escrito — solo **lee y entiende**:
> - ✅ La estructura `class Equipo:` y los tres `def`
> - ✅ `self.activo = False` — todo equipo comienza apagado
> - ✅ La lógica del método `estado()` ya está completa
>
> Debes **completar** — reemplaza cada `___`:
> - ✏️ `___` `self.nombre = ___` y `self.codigo = ___` en `__init__`
> - ✏️ `___` El `print` de `encender()` — f-string con `self.nombre` y `self.codigo`
> - ✏️ `___` El `print` de `apagar()` — mismo patrón que `encender()`


````python
class Equipo:
    def __init__(self, nombre, codigo):
        self.nombre = nombre              # guarda el parámetro nombre
        self.codigo = codigo              # guarda el parámetro codigo
        self.activo = False            # todo equipo comienza apagado

    def encender(self):
        self.activo = True
        print(f"[ON]  {self.nombre} ({self.codigo}) encendido.")    # nombre y codigo

    def apagar(self):
        self.activo = False
        print(f"[OFF] {self.nombre} ({self.codigo}) apagado.")      # mismo patrón

    def estado(self):
        condicion = "ACTIVO" if self.activo else "INACTIVO"
        print(f"Equipo: {self.nombre}  |  Estado: {condicion}")
````

Ejecuta la celda de arriba. Luego verifica que el sistema registra equipos correctamente:

````python
# ── VERIFICACIÓN BLOQUE 1 ────────────────────────────────
e1 = Equipo("Mezclador Central", "EQ-001")
e1.estado()        # debe decir INACTIVO
e1.encender()
e1.estado()        # debe decir ACTIVO
e1.apagar()

# Salida esperada:
# Equipo: Mezclador Central  |  Estado: INACTIVO
# [ON]  Mezclador Central (EQ-001) encendido.
# Equipo: Mezclador Central  |  Estado: ACTIVO
# [OFF] Mezclador Central (EQ-001) apagado.
````

# 🔷 BLOQUE 2 — EQUIPOS ESPECIALIZADOS — `Reactor` y `Bomba`

---

Revisas el cuaderno. Hay una hoja de especificaciones técnicas:

| Equipo | Código | Capacidad | Temp. máx |
|--------|--------|-----------|-----------|
| Reactor-A | RX-001 | 5 000 L | 180 °C |

| Equipo | Código | Caudal | Presión |
|--------|--------|--------|---------|
| Bomba Principal | BP-001 | 120 L/min | 8.5 bar |
| Bomba Auxiliar | BP-002 | 80 L/min | 5.0 bar |

Un reactor **ES** un equipo — tiene nombre, código, se puede encender —
pero además tiene datos propios. La solución: **herencia**.

---


```python
# ── Estado del sistema hasta aquí ──────────────────────
# [X] class Equipo              (ya definida — no la toques)
# [ ] class Reactor(Equipo)     ← construyes aquí
# [ ] class Bomba(Equipo)       ← construyes aquí
```

> 💡 **Concepto clave**
>
> **Herencia:** `Reactor(Equipo)` significa que `Reactor` **es** un `Equipo`. Hereda todo lo del padre — `nombre`, `codigo`, `encender()`, `apagar()` — y solo agrega lo nuevo. Sin repetir código.


### Regla de oro

```python
class ClaseHija(ClasePadre):
    def __init__(self, args_padre, args_nuevos):
        super().__init__(args_padre)      # ← SIEMPRE primero: activa al padre
        self.nuevo_atributo = nuevo_valor # ← luego agrega lo propio
```


> **Cómo leer esta celda:**
>
> Ya escrito — solo **lee y entiende**:
> - ✅ `class Reactor(Equipo):` — declaración de herencia
> - ✅ `super().__init__(nombre, codigo)` — activa el constructor del padre
> - ✅ La estructura de `info_reactor()` ya está dada
>
> Debes **completar** — reemplaza cada `___`:
> - ✏️ `___` Los dos parámetros nuevos del `__init__` de `Reactor`
> - ✏️ `___` `self.capacidad_L = ___` y `self.temp_max = ___`
> - ✏️ `___` Los valores en los `print` de `info_reactor()` — usa `self.`


````python
class Reactor(Equipo):                               # hereda de Equipo
    def __init__(self, nombre, codigo, capacidad_L, temp_max):    # 2 parámetros nuevos
        super().__init__(nombre, codigo)             # activa el constructor padre
        self.capacidad_L = capacidad_L                       # litros
        self.temp_max    = temp_max                       # grados Celsius

    def info_reactor(self):
        print(f"Reactor: {self.nombre}")
        print(f"  Capacidad : {self.capacidad_L} L")
        print(f"  Temp. máx : {self.temp_max} °C")

````

> **Cómo leer esta celda:**

````python
class Bomba(Equipo):                                 # hereda de Equipo
    def __init__(self, nombre, codigo, caudal_Lmin, presion_bar):    # 2 parámetros nuevos
        super().__init__(nombre, codigo)
        self.caudal_Lmin = caudal_Lmin                       # litros por minuto
        self.presion_bar = presion_bar                       # bar

    def info_bomba(self):
        print(f"Bomba: {self.nombre}")
        print(f"  Caudal  : {self.caudal_Lmin} L/min")
        print(f"  Presión : {self.presion_bar} bar")
````

# 🔷 BLOQUE 3 — SECUENCIA DE ENCENDIDO — Override
**⏱ Tiempo estimado: 12 min**

---

---
> ### 📋 SITUACIÓN
>
> Son las 12:30 PM. El operador de turno vuelve, esta vez más preocupado:

> *"El sistema anterior tenía algo importante que olvidé mencionar.
> Un reactor NO se enciende igual que un equipo genérico — primero hay que
> presurizarlo o puede haber un accidente. Y la bomba necesita que verifiques
> las válvulas antes. El sistema debe reflejar eso en el encendido."*

Miras tu código. El método `encender()` actual es igual para todos.
Necesitas que cada equipo tenga **su propia secuencia de encendido**
sin tocar la clase `Equipo` original.

La solución: **override** — redefinir el método en el hijo.

---


> ⚠️ **DEPENDENCIA** — Este bloque **redefine** `Reactor` y `Bomba`. Completa el TU TURNO del Bloque 2 antes de continuar — la verificación del Bloque 4 depende de que ambos overrides estén listos.


```python
# ── Estado del sistema hasta aquí ──────────────────────
# [X] class Equipo              (no la toques — es el padre)
# [~] class Reactor(Equipo)     ← agregas encender() adentro
# [~] class Bomba(Equipo)       ← tú completas encender() adentro
```

> 💡 **Concepto clave**
>
> **Override:** defines el **mismo nombre de método** en el hijo. Python usa la versión del hijo automáticamente. Si además quieres ejecutar la secuencia del padre, llamas a `super().encender()` — que activa el `[ON]` original.


### Flujo de trabajo de esta celda

1. **Ejecuta** la celda (con el `pass` en Bomba)
2. **Prueba** el TEST de Reactor — verifica que funciona
3. **Completa** el `___` de Bomba (reemplaza el `pass`)
4. **Re-ejecuta** la celda
5. **Prueba** el TEST de Bomba


> **Cómo leer esta celda:**
>
> Ya escrito — solo **lee y entiende**:
> - ✅ `Reactor.encender()` ya tiene el override completo — léelo como modelo
> - ✅ La estructura `def encender(self):` en `Bomba` ya está declarada
> - ✅ `super().encender()` en Bomba ya está escrito — no lo quites
>
> Debes **completar** — reemplaza cada `___`:
> - ✏️ `___` En `Bomba.encender()`: el `print` antes del `super()` — `'Verificando válvulas [nombre]...'`
> - ✏️ `___` En `Bomba.encender()`: el `print` después del `super()` — `'Caudal nominal: X L/min'`


````python
class Reactor(Equipo):
    def __init__(self, nombre, codigo, capacidad_L, temp_max):
        super().__init__(nombre, codigo)
        self.capacidad_L = capacidad_L
        self.temp_max    = temp_max

    def encender(self):                              # OVERRIDE del padre
        print(f"Presurizando {self.nombre}...")
        super().encender()                           # ejecuta el [ON] original
        print(f"Temp. objetivo: {self.temp_max} °C")

    def info_reactor(self):
        print(f"Reactor: {self.nombre}")
        print(f"  Capacidad : {self.capacidad_L} L")
        print(f"  Temp. máx : {self.temp_max} °C")


class Bomba(Equipo):
    def __init__(self, nombre, codigo, caudal_Lmin, presion_bar):
        super().__init__(nombre, codigo)
        self.caudal_Lmin = caudal_Lmin
        self.presion_bar = presion_bar

    def encender(self):                              # OVERRIDE — completa tú
        print(f"Verificando las valvulas de {self.nombre}...")               # "Verificando válvulas [nombre]..."
        super().encender()
        print(f"Caudal nomial {self.caudal_Lmin} L/min")                # "Caudal nominal: X L/min"

    def info_bomba(self):
        print(f"Bomba: {self.nombre}")
        print(f"  Caudal  : {self.caudal_Lmin} L/min")
        print(f"  Presión : {self.presion_bar} bar")
````

````python
# ── VERIFICACIÓN BLOQUE 3 ────────────────────────────────
print("--- Equipo genérico ---")
Equipo("Agitador", "EQ-010").encender()

print()
print("--- Reactor con secuencia de encendido ---")
Reactor("Reactor-A", "RX-001", 5000, 180).encender()

print()
print("--- Bomba con verificación de válvulas ---")
Bomba("Bomba Principal", "BP-001", 120, 8.5).encender()

# Salida esperada:
# --- Equipo genérico ---
# [ON]  Agitador (EQ-010) encendido.
#
# --- Reactor con secuencia de encendido ---
# Presurizando Reactor-A...
# [ON]  Reactor-A (RX-001) encendido.
# Temp. objetivo: 180 °C
#
# --- Bomba con verificación de válvulas ---
# Verificando válvulas Bomba Principal...
# [ON]  Bomba Principal (BP-001) encendida.
# Caudal nominal: 120 L/min
````

# 🔷 BLOQUE 4 — ARRANQUE DE PLANTA — Polimorfismo
**⏱ Tiempo estimado: 8 min**

---

---
> ### 📋 SITUACIÓN
>
> Son las 13:45 PM. El ingeniero jefe entra al cuarto de control con cara seria:

> *"Quince minutos para el turno de la tarde. ¿Podemos arrancar la planta?"*

El sistema tiene registrados: el Agitador, el Reactor-A, la Bomba Principal y el Reactor-B.
Cada uno con su propia secuencia de encendido.

El operador necesita **un solo comando** que encienda todos los equipos en orden,
sin importar qué tipo de equipo sea cada uno.

Tu código tiene que hacer eso — sin un `if` para cada tipo.
Eso es **polimorfismo**.

---


> ⚠️ **DEPENDENCIA** — Requiere que el override de `Bomba` del Bloque 3 esté completo y la celda re-ejecutada.


> 💡 **Concepto clave**
>
> **Polimorfismo:** el operador presiona el mismo botón `ENCENDER` para cualquier equipo. El reactor presuriza, la bomba verifica válvulas, el agitador solo enciende. El operador no sabe ni le importa el tipo exacto — solo sabe que todos responden a `encender()`. Python elige la versión correcta automáticamente en tiempo de ejecución.


### El poder del polimorfismo

```python
for equipo in planta:
    equipo.encender()    # Python elige la versión correcta — sin if/elif
```

La celda de abajo tiene el `for` loop listo. Solo debes **completar la lista `planta`**
con los cuatro equipos de la planta usando los datos del cuaderno.


> **Cómo leer esta celda:**
>
> Ya escrito — solo **lee y entiende**:
> - ✅ El `for` loop con `equipo.encender()` — no lo toques
> - ✅ Los comentarios indican qué objeto va en cada línea
>
> Debes **completar** — reemplaza cada `___`:
> - ✏️ `___` Los argumentos de cada objeto en la lista — usa los datos de la tabla del Bloque 2


````python
# ── ARRANQUE DE PLANTA — 13:58 PM ────────────────────────
planta = [
    Equipo("Agitador", "EQ-010"),
    Reactor("Reactor-A", "RX-001", 5000, 180),
    Bomba("Bomba Principal", "BP-001", 120, 8.5),
    Reactor("Reactor-B", "RX-002", 4000, 200),
]

print("=== ARRANQUE DE PLANTA — TURNO TARDE ===")
for equipo in planta:
    equipo.encender()
    print()

# Salida esperada:
# === ARRANQUE DE PLANTA — TURNO TARDE ===
# [ON]  Agitador (EQ-010) encendido.
#
# Presurizando Reactor-A...
# [ON]  Reactor-A (RX-001) encendido.
# Temp. objetivo: 180 °C
#
# Verificando válvulas Bomba Principal...
# [ON]  Bomba Principal (BP-001) encendida.
# Caudal nominal: 120 L/min
#
# Presurizando Reactor-B...
# [ON]  Reactor-B (RX-002) encendido.
# Temp. objetivo: 200 °C
````

### Conceptos aplicados

| Concepto | Qué hace | Sintaxis clave |
|----------|----------|----------------|
| **Clase** | Plano/ficha técnica para crear objetos | `class Nombre:` |
| **Constructor** | Inicializa los datos del objeto al crearlo | `def __init__(self, ...)` |
| **Herencia** | La hija reutiliza todo del padre y agrega lo suyo | `class Hija(Padre):` |
| **`super()`** | Llama al método del padre desde el hijo | `super().__init__(...)` |
| **Override** | El hijo redefine la secuencia de un método del padre | mismo `def metodo(self):` |
| **Polimorfismo** | Un solo comando activa todos los equipos — cada uno a su manera | `for obj in lista: obj.m()` |
